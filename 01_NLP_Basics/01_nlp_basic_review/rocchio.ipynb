```python
# Step B: Rocchio formula
# Purpose: define how the updated query is computed
# Code meaning: combine the original query, the relevant centroid, and the non-relevant centroid
# Result: a new query vector
```

```python
q0 = {"nlp": 1.0, "text": 1.0}

relevant_docs = [
    {"nlp": 2.0, "text": 1.0, "mining": 1.0},
    {"nlp": 1.0, "text": 1.0, "retrieval": 1.0}
]

nonrelevant_docs = [
    {"dogs": 1.0, "cute": 1.0},
    {"fun": 1.0, "dogs": 1.0}
]

print("Original query:", q0)
print("Relevant docs:", relevant_docs)
print("Non-relevant docs:", nonrelevant_docs)
```

```python
q0 = {"nlp": 1.0, "text": 1.0}

relevant_docs = [
    {"nlp": 2.0, "text": 1.0, "mining": 1.0},
    {"nlp": 1.0, "text": 1.0, "retrieval": 1.0}
]

nonrelevant_docs = [
    {"dogs": 1.0, "cute": 1.0},
    {"fun": 1.0, "dogs": 1.0}
]

print("Original query:", q0)
print("Relevant docs:", relevant_docs)
print("Non-relevant docs:", nonrelevant_docs)
```

```python
# Step D: Compute centroid
# Purpose: compute the average vector of a group of documents
# Code meaning: sum the term weights across documents and divide by the number of documents
# Result: a centroid vector
```

```python
def compute_centroid(docs):
    centroid_sum = {}

    for doc in docs:
        for term, weight in doc.items():
            if term not in centroid_sum:
                centroid_sum[term] = weight
            else:
                centroid_sum[term] += weight

    N = len(docs)

    for term in centroid_sum:
        centroid_sum[term] = centroid_sum[term] / N

    return centroid_sum
```

```python
cr = compute_centroid(relevant_docs)
cnr = compute_centroid(nonrelevant_docs)

print("Relevant centroid:", cr)
print("Non-relevant centroid:", cnr)
```

```python
# Step F: Rocchio update
# Purpose: compute the updated query vector
# Code meaning: apply the Rocchio formula to all terms
# Result: a new query vector with adjusted weights
```

```python
def rocchio_update(q0, cr, cnr, alpha=1.0, beta=0.75, gamma=0.15):
    updated = {}

    all_terms = set(q0.keys()) | set(cr.keys()) | set(cnr.keys())

    for term in all_terms:
        updated[term] = (
            alpha * q0.get(term, 0)
            + beta * cr.get(term, 0)
            - gamma * cnr.get(term, 0)
        )

    return updated
```

```python
qm = rocchio_update(q0, cr, cnr)

print("Updated query:", qm)
```

```python
#nlp and text get higher weights because they are in the original query and also appear in relevant documents.
#mining and retrieval are added because they appear in relevant documents.
#dogs, cute, and fun get negative weights because they come from non-relevant documents.
#This shows that Rocchio moves the query toward relevant documents and away from non-relevant documents.
```
