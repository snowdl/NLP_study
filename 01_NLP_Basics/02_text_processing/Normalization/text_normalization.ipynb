```python
import re
import html
import unicodedata

text = """
Hello!!!   My email is TEST123@example.com.
Visit https://test.com now.
가격은 12,000원입니다. <b>This is HTML</b>
Café — NLP is fun :)
"""

print("0. 원문")
print(text)
```

```python
# 1) lowercasing
lower_text = text.lower()
print("\n1. lowercasing")
print(lower_text)
```

```python
# 2) whitespace normalization
normalized_space = re.sub(r"\s+", " ", lower_text).strip()
print("\n2. whitespace normalization")
print(normalized_space)
```

```python
# 3) unicode normalization
normalized_unicode = unicodedata.normalize("NFKC", normalized_space)
print("\n3. unicode normalization")
print(normalized_unicode)
```

```python
#-------------------------
# B.Pattern-based cleaning
# -------------------------
```

```python
# 4) HTML removal

# re.sub(pattern, replacement, text)
# pattern: r"<[^>]+>"
#   <       : opening angle bracket
#   [^>]    : any character except ">"
#   +       : one or more times
#   >       : closing angle bracket
# So this pattern matches HTML tags such as <b>, </p>, <div>, etc.
# These tags are replaced with a single space " ".
no_html = re.sub(r"<[^>]+>", " ", normalized_unicode)

# html.unescape() converts HTML entities back to normal characters.
# Example:
#   &amp;  -> &
#   &lt;   -> <
#   &gt;   -> >
#   &quot; -> "
no_html = html.unescape(no_html)

# Print the result after HTML removal
print("\n4. HTML removal")
print(no_html)
```

```python
import re  # Import Python's regular expression module

# Original text that contains two URLs:
# 1) https://openai.com
# 2) www.google.com
no_html = "Visit https://openai.com now and also check www.google.com please."

# re.sub(pattern, replacement, string)
# This function searches the string for parts that match the regex pattern,
# and replaces each match with the replacement string.
#
# pattern: r"https?://\S+|www\.\S+"
#
# Let's break this pattern down:
#
# r" ... "
# - The r means "raw string".
# - It tells Python to treat backslashes (\) literally,
#   which is very useful when writing regex patterns.
#
# https?://\S+
# - http      : matches the characters "http"
# - s?        : matches "s" zero or one time
#               so it matches both "http" and "https"
# - ://       : matches the literal characters "://"
# - \S+       : matches one or more non-whitespace characters
#               \S means "not a space/tab/newline"
#               + means "one or more"
#
# So this part matches URLs like:
# - http://example.com
# - https://openai.com
# - https://openai.com/blog
#
# |
# - This means "or" in regex.
#
# www\.\S+
# - www       : matches the literal characters "www"
# - \.        : matches a literal dot "."
#               We use \. because a plain . in regex means "any character"
# - \S+       : matches one or more non-whitespace characters after "www."
#
# So this part matches URLs like:
# - www.google.com
# - www.naver.com
# - www.example.org/page
#
# replacement: " "
# - Every matched URL is replaced with a single space.
# - We use a space instead of deleting completely so words do not get stuck together.
#
# string: no_html
# - This is the original text we want to clean.
no_url = re.sub(r"https?://\S+|www\.\S+", " ", no_html)

# Print the cleaned text after URL removal
print(no_url)
```

```python
# 6) email removal

# Remove email addresses using regex
# \b              : word boundary
# [\w\.-]+        : one or more word characters, dots, or hyphens
# @               : literal @ symbol
# [\w\.-]+        : domain name part
# \.              : literal dot before the top-level domain
# \w+             : top-level domain such as com, org, net
# \b              : word boundary
# Replace matched email addresses with a single space
no_email = re.sub(r"\b[\w\.-]+@[\w\.-]+\.\w+\b", " ", no_url)

print("\n6. email removal")
print(no_email)
```
