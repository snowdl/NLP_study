```python
docs = [
    "I really like NLP, and I like text mining too! NLP is fun.",
    "Text mining is useful for NLP tasks.",
    "Dogs are fun and cute."
]
```

```python
import re
def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    tokens = text.split()
    return tokens
```

```python
docs_tokens = [preprocess(doc) for doc in docs]
print(docs_tokens)
```

```python
# Step J: Document Frequency (DF)
df = Counter()  # create an empty counter to store document frequency
for doc in docs_tokens:  # loop through each document
    for term in set(doc):  # use set(doc) so each term is counted only once per document
        df[term] += 1  # increase the document count for that term

print(df)  # show how many documents contain each term
```

```python
#IDF measures how rare a term is across the whole document collection.
#Key idea:
#If a term appears in many documents, its IDF is low.
#If a term appears in only a few documents, its IDF is high.
#In other words:
#Common words like is and and are less important.
#Terms that appear in only one or a few documents are more important.
```

```python
# Step K: Inverse Document Frequency (IDF)

import math

N = len(docs_tokens)  # total number of documents
idf = {}  # store IDF values

for term, freq in df.items():  # loop through each term and its document frequency
    idf[term] = math.log(N / freq)  # compute IDF using log(N / df)

print(idf)
```

```python
#Step L: TF-IDF
#Purpose: combine term frequency and inverse document frequency
#Code meaning: multiply TF by IDF for each term
#Result: terms that are frequent in a document but rare across documents get higher scores
```

```python
tfidf_docs = []

for tf in tf_docs:  # loop through each document's TF dictionary
    tfidf = {}  # store TF-IDF values for one document
    for term, freq in tf.items():  # loop through each term and its TF value
        tfidf[term] = freq * idf[term]  # compute TF-IDF
    tfidf_docs.append(tfidf)  # add this document's TF-IDF to the list

print(tfidf_docs)
```

```python
# Step M: Basic inverted index
# Purpose: map each term to the documents that contain it
# Code meaning: for each document, add each unique term to the index with its document ID
# Result: each term points to a postings list of document IDs
#term마다그 term가 등장한 문서 번호들을 모아서리스트로 저장한 것
```

```python
#term -> postings list
#'dogs' -> [3]
```

```python
inverted_index = {}  # store term -> list of document IDs

for doc_id, doc in enumerate(docs_tokens, start=1):  # loop through each document with its ID
    for term in set(doc):  # count each term only once per document
        if term not in inverted_index:  # if the term is not in the index yet
            inverted_index[term] = [doc_id]  # create a new postings list
        else:
            inverted_index[term].append(doc_id)  # add the document ID to the postings list

print(inverted_index)
```

```python
# Step N: Postings list with TF
# Purpose: store both document IDs and term frequency for each term
# Code meaning: for each document, add (doc_id, tf) for every term
# Result: each term points to a postings list with document IDs and TF values
#어느 문서에 있고, 그 문서 안에서 몇 번 나왔는지도 보여줌
```

```python
postings_tf = {}  # store term -> list of (doc_id, tf)

for doc_id, tf in enumerate(tf_docs, start=1):  # loop through each document's TF dictionary
    for term, freq in tf.items():  # loop through each term and its frequency
        if term not in postings_tf:  # if the term is not in the postings list yet
            postings_tf[term] = [(doc_id, freq)]  # create a new postings list
        else:
            postings_tf[term].append((doc_id, freq))  # add (doc_id, tf) to the postings list

print(postings_tf)
```

```python
#Step O: Champion list
#Purpose: keep only the top-ranked documents for each term
#Code meaning: sort each postings list by a score such as TF or TF-IDF, then keep the top documents
#Result: each term points to a shorter list of high-scoring documents
```

```python
r = 1  # keep top 1 document for each term

champion_lists = {}

for term, postings in postings_tf.items():  # loop through each term and its postings list
    sorted_postings = sorted(postings, key=lambda x: x[1], reverse=True)  # sort by TF
    champion_lists[term] = sorted_postings[:r]  # keep only the top r postings

print(champion_lists)
```

```python
# Step P-a: Query preparation
#preprocessing = 텍스트를 정리하는 전체 과정
```

```python
query = "nlp text mining"
print(query)
```

```python
query_tokens = preprocess(query)
print(query_tokens)
```

```python
# Step R: Query TF
# Purpose: count how often each term appears in the query
# Code meaning: use Counter to compute term frequency in the query
# Result: a TF dictionary for the query
```

```python
from collections import Counter

query_tf = Counter(query_tokens)
print(query_tf)
```

```python
# Step S: Query TF-IDF
# Purpose: convert the query into a TF-IDF vector
# Code meaning: multiply query TF by the IDF value of each term
# Result: a TF-IDF dictionary for the query
```

```python
query_tfidf = {}

for term, freq in query_tf.items():
    query_tfidf[term] = freq * idf.get(term, 0)

print(query_tfidf)
```

```python
# Step T: Cosine similarity function
# Purpose: measure how similar two TF-IDF vectors are
# Code meaning: compute dot product and divide by vector lengths
# Result: a similarity score between 0 and 1
```

```python
import math

def cosine_dict(vec1, vec2):
    # get all terms that appear in either vector
    all_terms = set(vec1.keys()) | set(vec2.keys())

    # compute dot product
    dot_product = 0
    for term in all_terms:
        dot_product += vec1.get(term, 0) * vec2.get(term, 0)

    # compute norm of vec1
    norm1 = 0
    for value in vec1.values():
        norm1 += value ** 2
    norm1 = math.sqrt(norm1)

    # compute norm of vec2
    norm2 = 0
    for value in vec2.values():
        norm2 += value ** 2
    norm2 = math.sqrt(norm2)

    # avoid division by zero
    if norm1 == 0 or norm2 == 0:
        return 0

    return dot_product / (norm1 * norm2)
```

```python
# Step U: Query-document cosine similarity
# Purpose: compare the query with each document
# Code meaning: compute cosine similarity between the query TF-IDF vector and each document TF-IDF vector
# Result: documents can be ranked by relevance
```

```python
for i, doc_tfidf in enumerate(tfidf_docs, start=1):
    score = cosine_dict(query_tfidf, doc_tfidf)
    print(f"D{i} cosine similarity:", round(score, 4))
```

```python
# Step V: Ranking
# Purpose: rank documents by similarity to the query
# Code meaning: sort documents by cosine similarity score in descending order
# Result: the most relevant document appears first
```

```python
scores = []

for i, doc_tfidf in enumerate(tfidf_docs, start=1):
    score = cosine_dict(query_tfidf, doc_tfidf)
    scores.append((i, score))

ranked_scores = sorted(scores, key=lambda x: x[1], reverse=True)
print(ranked_scores)
```
