```python
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
from IPython.display import display, HTML
POS_COLORS = {
    'NN': 'blue',     # nouns
    'VB': 'red',      # verbs
    'JJ': 'green',    # adjectives
    'RB': 'purple',   # adverbs
}


def color_pos(tagged_sentence):
    html_parts = []
    for token, tag in tagged_sentence:
        prefix = tag[:2]   # NN, VB, JJ, RB
        color = POS_COLORS.get(prefix, 'black')
        html_parts.append(f"<span style='color:{color}'>{token}({tag})</span>")
    display(HTML(" ".join(html_parts)))
```

```python
def pos_tag_sentence(sentence):
    tokens = word_tokenize(sentence)
    return pos_tag(tokens)


def run_pos_experiment(sentences):
    results = {}
    for sent in sentences:
        results[sent] = pos_tag_sentence(sent)
    return results


def inspect(results):
    for sent, tagged in results.items():
        print(f"\nSentence: {sent}")
        color_pos(tagged) 


sentences = [
    "How much Chinese silk was exported to Western Europe by the end of the 18th century?"
]

results = run_pos_experiment(sentences)
inspect(results)
```
