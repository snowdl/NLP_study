```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

#np.exp(0) = 1
#입력이 커지면 (x → +∞), np.exp(-x) → 0 → sigmoid(x) → 1에 가까워짐
#입력이 작아지면 (x → -∞), np.exp(-x) → ∞ → sigmoid(x) → 0에 가까워짐
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
```

```python
plt.plot(x,y,'g')
plt.plot([0, 0], [1.0, 0.0], ':')
plt.title('sigmoid Function')
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))


x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)


plt.plot(x, y1, 'r', linestyle='--') # the value of w = 0.5
plt.plot(x, y2, 'g') # the value of w = 1
plt.plot(x, y3, 'b', linestyle='--') # the value of w = 2
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()

```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))


x = np.arange(-5.0, 5.0, 0.1)
#when b(=bias) added
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)


plt.plot(x, y1, 'r', linestyle='--') # the value of w = 0.5
plt.plot(x, y2, 'g') # the value of w = 1
plt.plot(x, y3, 'b', linestyle='--') # the value of w = 2
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
#As the bias b increases, the sigmoid function shifts to the left (a horizontal shift along the x-axis relative to x = 0).
```
