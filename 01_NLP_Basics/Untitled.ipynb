```python
def line_splitting(lines):
    out = []
    for s in lines:
        # 줄 끝 개행/양쪽 공백 제거 후, 왼쪽에서 한 번만 자르기
        left, right = s.strip().split(',', 1)
        # 각 조각도 공백 정리 (쉼표 뒤에 공백 있을 수 있음)
        out.append([left.strip(), right.strip()])
    return out
```

```python
inp = ["Smith,Cathryn,40", "  Doe, John,28  ", "Lee,Min,35"]
print(line_splitting(inp))
# 기대: [['Smith', 'Cathryn,40'], ['Doe', 'John,28'], ['Lee', 'Min,35']]
```

```python
def line_splitting(lines):
    """Return a list of tuples split by commas for each line."""
    result = []
    for s in lines:
        parts = s.split(',')        # str.split() 사용
        result.append(tuple(parts)) # 각 줄을 튜플로
    return result
```

```python
inp = ["Smith,Cathryn,40", "  Doe, John,28  ", "Lee,Min,35"]
print(line_splitting(inp))
# 기대: [['Smith', 'Cathryn,40'], ['Doe', 'John,28'], ['Lee', 'Min,35']]
```

```python
def line_splitting(lines):
    ''' Take one or more lines as input, and split each line
        based on commas

        input: lines (a list of strings)
        returns: a list of tuples
    '''
    result = []
    for s in lines:
        parts = s.split(',')          # 쉼표로 분리
        result.append(tuple(parts))   # 튜플로 변환
    return result
```

```python
inp = ["Smith,Cathryn,40", "  Doe, John,28  ", "Lee,Min,35"]
print(line_splitting(inp))
# 기대: [['Smith', 'Cathryn,40'], ['Doe', 'John,28'], ['Lee', 'Min,35']]
```
