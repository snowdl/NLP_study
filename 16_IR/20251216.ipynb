```python
from collections import Counter
import math
```

```python
# Train data (labels + token lists)
# 목표 카운트:
# Positive: good=4, nice=2, bad=0 (total=6), docs=3
# Negative: good=1, nice=0, bad=3 (total=4), docs=2

train_data = [
    ("Positive", ["good", "good"]),        # good 2
    ("Positive", ["good", "nice"]),        # good 1, nice 1
    ("Positive", ["good", "nice"]),        # good 1, nice 1  -> total good=4, nice=2

    ("Negative", ["bad", "bad"]),          # bad 2
    ("Negative", ["bad", "good"]),         # bad 1, good 1   -> total bad=3, good=1
]

# Test doc (우리가 손계산한 그 문서)
test_doc = ["good", "bad"]
```

```python
#Create priors

labels = [label for label, words in train_data]
```

```python
print(labels)
```

```python
label_counts = Counter(labels)
```

```python
print(label_counts)
```

```python
total_docs = len(train_data)
```

```python
print(total_docs)
```

```python
def compute_priors(label_counts, total_docs):
    priors = {}
    for label in label_counts:
        priors[label] = label_counts[label] / total_docs
    return priors

priors = compute_priors(label_counts, total_docs)
print(priors)
```

```python
#conditional probability
```

```python
#분자
word_counts = {
    "Positive" : Counter(),
    "Negative" : Counter()
}

#분모
total_words ={
    "Positive":0,
    "Negative":0
    
}
```

```python
for label, words in train_data:
    for w in words :
        word_counts[label][w] += 1
        total_words[label] += 1


print(word_counts)
print(total_words)
```

```python
#vocab size
```

```python
vocab = {w for label, words in train_data for w in words}

vocab_size = len(vocab)

print("vocab :", vocab)
print("vocab_size :",vocab_size)
```

```python
#prediction with laplace smoothing +1
```

```python
def predict_nb(test_doc_tokens):
    scores = {}

    for label in priors:
        score = math.log(priors[label])

        # likelihood (Laplace smoothing + log)
        for w in test_doc_tokens:
            count_wc = word_counts[label][w] 
            prob = (count_wc + 1) / (total_words[label] + vocab_size)
            score += math.log(prob)

        scores[label] = score  # 1

    predicted = max(scores, key=scores.get)
    return predicted, scores
 
```

```python
pred, scores = predict_nb(["good", "bad"])
print(pred)
print(scores)
```
