```python
from collections import Counter
import math
```

```python
train_data = {
    "Positive": [
        ["good", "good", "good"],
        ["good", "good"],
        ["good"],
        ["good", "bad"],
        ["bad"]
    ],
    "Negative": [
        ["good", "good"],
        ["good"],
        ["good", "bad"],
        ["bad", "bad"],
        ["bad", "bad"],
        ["bad"],
        ["bad"]
    ]
}
```

```python
#전체 문서 구하기 
total_docs = 0
for label, docs in train_data.items():
    total_docs += len(docs)

print("total_docs:", total_docs)
```

```python
#Prior per class
```

```python
priors = {}

for label, docs in train_data.items():
    priors[label] = len(docs) / total_docs

print(priors)
```

```python
#log priors: likelihood랑 같은 스케일(log) 로 맞추기 위함
import math

log_priors = {}

for label, p in priors.items():
    log_priors[label] = math.log(p)

print(log_priors)
```

```python
#분자, 클래스별 단어 카운트
word_counts = {
    "Positive": Counter(),
    "Negative": Counter()
}

#분모/ 클래스별 단어 수 저장
total_words = {
    "Positive": 0,
    "Negative": 0
}
```

```python
total_docs = len(train_data)
```

```python
print(total_docs)
```

```python
#word_counts 
```

```python
for label, docs in train_data.items():#train_data.items()에서 word까지 얻을 수 없기 때문
    for doc in docs:
        for word in doc:
            word_counts[label][word] += 1
            total_words[label] += 1
```

```python
print(word_counts)
```

```python
print(total_words)
```

```python
#vocab 
```

```python
vocab ={}
vocab = {w for label, docs in train_data.items() for doc in docs for w in doc}
```

```python
print(vocab)
```

```python
#vocab size
vocab_size = len(vocab)
print(vocab_size)
```

```python
def prediction_nb(text_doc):
    # 1) 클래스별 점수(log prior로 초기화)
    scores = {label: math.log(priors[label]) for label in priors}

    # 2) 각 단어에 대해 클래스별로 log likelihood 누적
    for word in text_doc:
        for label in scores:
            # count(w, c)
            count_wc = word_counts[label][word] if word in vocab else 0

            # P(w|c) with Laplace smoothing (alpha = 1)
            prob = (count_wc + 1) / (total_words[label] + vocab_size)

            # log P(w|c) 누적
            scores[label] += math.log(prob)

    # 3) 가장 점수가 큰 label 반환
    return scores
```

```python
text_doc = ["good", "bad", "bad"]
```

```python
prediction_nb(text_doc)
```
