```python
X = df.drop('price', axis=1)
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#MinMaxScaler--> scales the data so that minimum becomes 0 and maximum becomes 1.
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
#Sequential--> to create a linear stack of layers, where each layer has exactly one input tensor and one output tensor.
#Dense -->neuron connects to all neurons in the previous layer.
#Dense(units=number_of_neurons, activation='activation_function') 


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
# Dense(19, activation='relu')  4
#Four fully connected (dense) layers, each with 19 neurons.
#Using ReLU activation function to introduce non-linearity.
#Each layer learns more complex features from the previous layers output.


model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

#Dense(1)
#Final output layer with 1 neuron.
#No activation function (linear), suitable for regression tasks predicting continuous values.

model.add(Dense(1))


# compile(optimizer='adam', loss='mse')
#Optimizer: Adam (adaptive learning rate optimizer)
#Loss function: Mean Squared Error (MSE), commonly used for regression problems.


model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test),
         batch_size=128, epochs= 400)
```

```python
#model.history.history -> stores the training history logs (like loss, validation loss) as a dictionary

losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
#mean_squared_error (MSE)-->Average of squared differences between predicted and true values
#MSE->Penalizes larger errors more heavily

#mean_absolute_error (MAE)--> Average of absolute differences between predicted and true values
#MAE->Intuitive measure of average error magnitude


#explained_variance_score -->Measures how well the model explains the variance in the data
# Closer to 1 means better explanation; 0 means no explanation



from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error (y_test, predictions))
```
