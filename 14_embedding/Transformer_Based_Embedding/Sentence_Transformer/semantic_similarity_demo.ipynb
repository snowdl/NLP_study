## Purpose
The purpose of this practice is to learn how sentence embeddings work and how they can be used to measure semantic similarity between sentences.

## Tasks
- Convert sentences into embeddings
- Compare sentence meanings using cosine similarity
- Check whether semantically related sentences have higher similarity scores

## Learning Focus

## What this notebook does
In this notebook, I will:
1. Load a multilingual Sentence Transformer model
2. Prepare English and Korean sentences
3. Encode the sentences into embeddings
4. Compute cosine similarity
5. Observe which sentences are semantically close to each other
6. Find the most similar pair of sentences

```python
#import
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
```

```python
#prepare korean and english sentences
```

```python
sentences = [
    "I am studying NLP and information retrieval.",
    "I am learning about search and embeddings.",
    "My dog is sleeping on the sofa.",
    "나는 NLP와 정보검색을 공부하고 있다.",
    "나는 검색과 임베딩을 배우고 있다."
]

sentences
```

```python
#Step 1: Load a multilingual Sentence Transformer model
#I use a multilingual Sentence Transformer model so that I can compare both English and Korean sentences.
```

```python
model = SentenceTransformer("distiluse-base-multilingual-cased-v1")
```

```python
#Step 2 :Encode the sentences into embeddings
#Each sentence is converted into a numerical vector representation.
#These vectors are called sentence embeddings.
```

```python
embeddings = model.encode(sentences, convert_to_tensor=True)
embeddings.shape
```

```python
## Step 3. Compute cosine similarity
#This step compares all sentences with all other sentences.

#The result is a similarity matrix:
#- rows = sentences
#- columns = sentences
#- each value = cosine similarity score
```

```python
similarities = util.cos_sim(embeddings, embeddings)
similarities
```

```python
#Step 6. Observe which sentences are semantically close to each other
```

```python
df = pd.DataFrame(
    similarities.cpu().numpy(),
    index=sentences,
    columns=sentences
)

df
```

```python
# Step 7. Find the most similar pair of sentences
#The similarity matrix already contains all pairwise comparisons
```

```python
# Step 1. Set the initial best score
best_score = -1

# Step 2. Set the initial best pair
best_pair = None

# Step 3. Loop through the first sentence index
for i in range(len(sentences)):
    # Step 4. Loop through the second sentence index
     # Start from i+1 to avoid self-comparison and duplicate pairs
    for j in range(i + 1, len(sentences)):  # exclude self-comparison and duplicate pairs
       #Step 5. Get the similarity score for the pair (i, j)
        score = similarities[i][j].item()
         # Step 6. Check whether the current pair is better than the best pair so far
        if score > best_score:
            # Step 7. Update the best score
            best_score = score
            # Step 8. Update the best pair
            best_pair = (sentences[i], sentences[j])

# Step 9. Print the final result
print("Most similar pair:")
print(best_pair[0])
print(best_pair[1])
print(f"Similarity score: {best_score:.4f}")
```
