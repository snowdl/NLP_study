```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def relu(x):
    return np.maximum(0,x)

# Define ReLU (Rectified Linear Unit) activation function
def relu(x):
    return np.maximum(0, x)

# Create input values from -5.0 to 5.0 with step size 0.1
x = np.arange(-5.0, 5.0, 0.1)

# Apply ReLU function to the input
y = relu(x)

# Plot the ReLU function
plt.plot(x, y)
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
#Leaky Relu
'''
Allows a small, non-zero gradient when the input is negative, typically controlled by a small coefficient (e.g., 0.01).

Helps prevent the “dying ReLU” problem where neurons become inactive and stop learning.

Maintains the simplicity and efficiency of ReLU while improving gradient flow for negative inputs.
'''
```

```python
# Define the Leaky ReLU activation function with parameter a
def leaky_relu(x, a=0.01):
    return np.maximum(a * x, x)

# Generate input values
x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)  # Use default a=0.01

# Plot the function
plt.plot(x, y, label="Leaky ReLU")
plt.plot([0, 0], [5.0, -0.5], ":", color="gray")  # Vertical dotted line at x=0
plt.title("Leaky ReLU Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#softmax
'''
Converts a vector of real numbers into a probability distribution where all values sum to 1.

Commonly used in the output layer of multi-class classification models.

Emphasizes the largest input values while suppressing smaller ones, making the prediction interpretable as probabilities.

Sensitive to input scale; applying a numerical stability trick (like subtracting max input) is common to avoid overflow.
'''
```

```python
# Generate input values from -5.0 to 5.0 with step size 0.1
x = np.arange(-5.0, 5.0, 0.1)

# Compute softmax values: exponentiate x and normalize by sum of exponentials
y = np.exp(x) / np.sum(np.exp(x))

# Plot the softmax function
plt.plot(x, y)
plt.title('Softmax Function')
plt.xlabel('Input')
plt.ylabel('Output (Probability)')
plt.grid(True)
plt.show()
```

```python
#ELU(Exponential Linear Unit)
'''
Mitigates the "dying neuron" problem of ReLU in the region where inputs are less than or equal to zero.

Provides smooth transitions in the negative input region.

The alpha (α) parameter controls the slope for the negative region.
'''
```

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
x = np.linspace(-5, 5, 500)
y = elu(x)

plt.plot(x, y, label="ELU", color="magenta")
plt.title("ELU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

```python
'''
A smooth, non-monotonic activation function proposed by Google.
Exhibits better performance than ReLU in some deep networks due to smoothnes
'''
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def swish(x):
    return x * sigmoid(x)
```

```python
x = np.linspace(-5, 5, 500)
y = swish(x)

plt.plot(x, y, label="Swish", color="teal")
plt.title("Swish Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#GELU
'''
Gaussian Error Linear Unit, used widely in state-of-the-art models like BERT.

Combines properties of dropout and ReLU by smoothly gating inputs.
'''
```

```python
from scipy.stats import norm
```

```python
def gelu(x):
    return x * norm.cdf(x)

x = np.linspace(-5, 5, 500)
y = gelu(x)
```

```python
plt.plot(x, y, label="GELU", color="brown")
plt.title("GELU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define activation functions
def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, a=0.01):
    return np.maximum(a * x, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def softmax(x):
    exps = np.exp(x - np.max(x))  # Numerical stability
    return exps / np.sum(exps)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def swish(x):
    return x * sigmoid(x)

def gelu(x):
    return x * norm.cdf(x)

# Input range
x = np.linspace(-5, 5, 500)

# Calculate outputs
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_softmax = softmax(x)
y_elu = elu(x)
y_swish = swish(x)
y_gelu = gelu(x)

# Plot all functions
plt.figure(figsize=(14, 10))
plt.plot(x, y_relu, label='ReLU', color='red', linewidth=2)
plt.plot(x, y_leaky_relu, label='Leaky ReLU', color='orange', linewidth=2)
plt.plot(x, y_sigmoid, label='Sigmoid', color='blue', linewidth=2)
plt.plot(x, y_tanh, label='Tanh', color='green', linewidth=2)
plt.plot(x, y_softmax, label='Softmax', color='purple', linewidth=2)
plt.plot(x, y_elu, label='ELU', color='magenta', linewidth=2)
plt.plot(x, y_swish, label='Swish', color='teal', linewidth=2)
plt.plot(x, y_gelu, label='GELU', color='brown', linewidth=2)

plt.title('Comparison of Activation Functions', fontsize=18)
plt.xlabel('Input', fontsize=14)
plt.ylabel('Output', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
